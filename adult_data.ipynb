```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
columns_name=('age','workclass','education','education-num','marital-status','occupation','relationship','race','sex',
           'capital-gain','capital-loss','hours-per-week','native-country','target')
```

```python
df=pd.read_csv("data.csv",header=None)
```

```python
df.columns=columns_name
```

```python
df["age"]=df["age"].str.replace(",","")
```

```python
df["age"]=df["age"].astype("int")
```

```python
df['workclass']=df['workclass'].str.replace(",","")
df['education']=df['education'].str.replace(",","")
df['education-num']=df['education-num'].str.replace(",","")
df['marital-status']=df['marital-status'].str.replace(",","")
df['occupation']=df['occupation'].str.replace(",","")
df['relationship']=df['relationship'].str.replace(",","")
df['race']=df['race'].str.replace(",","")
df['sex']=df['sex'].str.replace(",","")
df['capital-gain']=df['capital-gain'].str.replace(",","")
df['capital-loss']=df['capital-loss'].str.replace(",","")
df['hours-per-week']=df['hours-per-week'].str.replace(",","")
df['native-country']=df['native-country'].str.replace(",","")
```

```python
df.columns
```

```python
df["workclass"]=df["workclass"].astype("int")
df["education-num"]=df["education-num"].astype("int")
df["capital-gain"]=df["capital-gain"].astype("int")
df["capital-loss"]=df["capital-loss"].astype("int")
df["hours-per-week"]=df["hours-per-week"].astype("int")
```

```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold',
alpha=0.8, y=1.)
for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold',
alpha=0.8, y=1.)
category = [ 'education', 'marital-status']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
